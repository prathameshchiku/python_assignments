### 39. Check if the sum of digits of a number is divisible by 3.

### 40. Check if a number is Armstrong (3-digit).

```python
n = 153
n_temp = str(n)
s = 0
for i in range(len(n_temp)):
    s += int(n_temp[i])**len(n_temp)
if s==n:
    print(f"{n} is an Armstrong Number")
else:
    print(f"{n} is Not an Armstrong Number")
```

### 42. Check if a number is perfect (sum of factors equals number).

```python
n = 28
s = 0
for i in range(1, n):
    if n%i==0:
        s += i

if s==n:
    print(f"{n} is a Perfect Number")
else:
    print(f"{n} is Not a Perfect Number")
```
