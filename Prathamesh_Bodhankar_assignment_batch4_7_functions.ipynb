### What does *args mean in Python? Write a function that accepts any number of values using *args and prints them.

```python
# Special syntax used in a function definition to allow the function to accept a variable number of non-keyword (positional) arguments.
def pnt(*numbers):
    for i in numbers:
        print(i)
pnt(44,4,46,46,4,46,41,31,13,1)
```

### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
l = [{"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5}, {"op":"sub","a": 65,"b": 59}, {"op":"div","a":59,"b":12}]
j = 1
def opertionss(l,j):
    for i in l:
        if i["op"]=="add":
            print(j, " : ", i["a"]+i["b"])
            j+=1
        if i["op"]=="sub":
            print(j, " : ", i["a"]-i["b"])
            j+=1
        if i["op"]=="mul":
            print(j, " : ", i["a"]*i["b"])
            j+=1
        if i["op"]=="div":
            print(j, " : ", i["a"]/i["b"])
            j+=1
opertionss(l,j)
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
l1 = [{"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[23,74,60]},{"name":"C", "marks":[88,99,77]}, {"name":"D", "marks":[69,77,66]},{"name":"E", "marks":[12,89,98]}, {"name":"F", "marks":[74,59,23]}] 
def process_students(l1):
    for i in l1:
        s = 0
        g = ""
        for j in i["marks"]:
            s += j
        if s>=270:
            g = "A"
        if s>=240 and s<270:
            g = "B"
        if s>=210 and s<240:
            g = "C"
        if s>=180 and s<210:
            g = "D"
        if s<180:
            g = "F"
        print("Name : ",i["name"]," | Total : ",s," | Grade : ",g)
process_students(l1)
```

### Create a function apply_discounts(products) that updates price after discount.

### Write a function average_salary(emp_list) that returns the average salary.

```python
l3 = [ {"id": 101, "name": "Amit", "age": 28, "department": "IT", "salary": 55000}, {"id": 102, "name": "Neha", "age": 24, "department": "HR", "salary": 40000}, {"id": 103, "name": "Raj", "age": 31, "department": "Finance", "salary": 65000}, {"id": 104, "name": "Simran", "age": 29, "department": "IT", "salary": 60000}, {"id": 105, "name": "Vijay", "age": 35, "department": "Admin", "salary": 35000} ]
def average_salary(l3):
    s = 0
    for i in l3:
        s += i['salary']
    return s/len(l3)
average_salary(l3)
```
