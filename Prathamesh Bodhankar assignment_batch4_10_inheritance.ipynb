# Python OOP Assignment — Inheritance Examples

```python
# 1) Single Inheritance: Person -> Student
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no

# TODO:
# 1. Create object of Student.
obj = Student("Abhijeet",1)
# 2. Print name.
print(obj.name)
# 3. Print roll_no.
print(obj.roll_no)
```

```python
# 2) Single Inheritance: Animal -> Dog
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# TODO:
# 1. Create Dog object.
obj = Dog()
# 2. Call sound() from parent class.
Animal().sound()
# 3. Call bark() from child class.
obj.bark()
```

```python
# 3) Constructor Inheritance: Vehicle -> Car
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

# TODO:
# 1. Create Car object.
obj = Car("Toyota","Corolla")
# 2. Print brand.
print(obj.brand)
# 3. Print model.
print(obj.model)
```

```python
# 4) Multilevel Inheritance: A -> B -> C
class A:
    def showA(self):
        print("This is A")

class B(A):
    def showB(self):
        print("This is B")

class C(B):
    def showC(self):
        print("This is C")

# TODO:
# 1. Create C object.
obj = C()
# 2. Call showA(), showB(), showC().
obj.showA()
obj.showB()
obj.showC()
```

```python
# 5) Multilevel Inheritance with methods: LivingThing -> Animal -> Dog
class LivingThing:
    def living(self):
        print("I am alive")

class Animal(LivingThing):
    def eat(self):
        print("Eating…")

class Dog(Animal):
    def bark(self):
        print("Barking…")

# TODO:
# 1. Create Dog object.
obj = Dog()
# 2. Call living(), eat(), bark().
obj.living()
obj.eat()
obj.bark()
```

```python
# 6) Multiple Inheritance: Father + Mother -> Child
class Father:
    def father_method(self):
        print("Father method")

class Mother:
    def mother_method(self):
        print("Mother method")

class Child(Father, Mother):
    def child_method(self):
        print("Child method")

# TODO:
# 1. Create Child object.
obj = Child()
# 2. Call father_method(), mother_method(), child_method().
obj.father_method()
obj.mother_method()
obj.child_method()
```

```python
# 7) Constructor Inheritance: Product -> Laptop
class Product:
    def __init__(self, name):
        self.name = name

class Laptop(Product):
    def __init__(self, name, ram):
        super().__init__(name)
        self.ram = ram

# TODO:
# 1. Create Laptop object.
obj = Laptop("Lenovo","128GB")
# 2. Print name.
print(obj.name)
# 3. Print ram.
print(obj.ram)
```

```python
# 8) Multilevel with __init__: Country -> State -> City
class Country:
    def __init__(self, cname):
        self.cname = cname

class State(Country):
    def __init__(self, cname, sname):
        super().__init__(cname)
        self.sname = sname

class City(State):
    def __init__(self, cname, sname, cityname):
        super().__init__(cname, sname)
        self.cityname = cityname

# TODO:
# 1. Create City object.
obj = City("India","Maharashtra","Pune")
# 2. Print cname, sname, cityname.
print(obj.cname)
print(obj.sname)
print(obj.cityname)
```

```python
# 9) Multiple Inheritance with methods: Teacher + Player -> Student
class Teacher:
    def teach(self):
        print("Teaching…")

class Player:
    def play(self):
        print("Playing…")

class Student(Teacher, Player):
    def study(self):
        print("Studying…")

# TODO:
# 1. Create Student object.
obj = Student()
# 2. Call teach(), play(), study().
obj.teach()
obj.play()
obj.study()
```

```python
# 10) Constructor Inheritance: Shape -> Rectangle
class Shape:
    def __init__(self, color):
        self.color = color

class Rectangle(Shape):
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

# TODO:
# 1. Create Rectangle object.
obj = Rectangle("Blue",24,25)
# 2. Print color, length, width.
print(obj.color)
print(obj.length)
print(obj.width)
```
