## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here
l = [[10,20], [30], [5, 7, 8]]
s = 0
for i in l:
    j = i
    for k in j:
        s += k
s
````

## **3. Count how many even numbers are present in a nested list.**

````python
# Write your answer here
l = [[10,20], [30], [5, 7, 8]]
c = 0
for i in l:
    j = i
    for k in j:
        if k%2==0:
            c += 1
c
````

## **4. Convert all strings to uppercase inside a nested list.**
Example: `[['hello', 'world'], ['python']]`

````python
# Write your answer here
l = [['hello', 'world'], ['python']]
for i in l:
    j = i
    for k in j:
        print(k.upper())
````

## **5. For the nested marks list, print total marks of each student.**
`[[80, 90, 85], [75, 70, 65], [88, 92, 79]]`

````python
# Write your answer here
l = [[80, 90, 85], [75, 70, 65], [88, 92, 79]]
n = 0
for i in l:
    j = i
    s = 0
    for k in j:
        s += k
    n += 1
    print(f"Student {n} Marks: {s}")
````

## **19. Find total stock by summing all 'quantity' values in a nested dictionary.**

````python
# Write your answer here
stock_data = {
    "warehouse_A": {
        "price": 0.8, "quantity": 200
    },
    "warehouse_B": {
        "price": 2.5, "quantity": 100
    },
    "warehouse_C": {
        "price": 3.0, "quantity": 50
    }
}

s = 0
for i,k in stock_data.items():
    s += k["quantity"]
print("Total Stock:",s)
````
